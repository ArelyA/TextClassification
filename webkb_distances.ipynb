```
import os
import numpy as np
os.sys.path.insert(0, os.path.join('/content/gdrive/My Drive/libraries/naive_bayes_vectorizer/'))
os.sys.path.insert(0, os.path.join('/content/gdrive/My Drive/libraries/utils/'))
os.sys.path.insert(0, os.path.join('/content/gdrive/My Drive/libraries/svm/'))
from utils import split, evaluate, save_distances
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nb_vectorizer import GaussianNB
from naive_bayes_vectorizer import ComplementNB
from svc import SVC
from scipy.spatial import distance

#classes = ['course', 'faculty', 'project', 'student']

currentFolder ='/content/gdrive/My Drive/'
c = [1.0, 10.0, 100.0, 1000.0, 10000.0, 100000.0]
kernel = ["linear", "poly", "rbf", "sigmoid"] 
```

```
# To get Counts
Count_x = CountVectorizer(max_features=75000)

# To vectorize
TFIDF_x = TfidfVectorizer(max_features=75000)
VectGaussian_NB = GaussianNB()
VectComplement_NB = ComplementNB()
```

```
dataset_test = open('/content/gdrive/My Drive/Datasets/webkb-test-stemmed.txt').read().split('\n')
dataset_train = open('/content/gdrive/My Drive/Datasets/webkb-train-stemmed.txt').read().split('\n')

X_train, y_train = split(dataset_train)
X_test, y_test = split(dataset_test)
```

```
#TFIDF
X_trainT = TFIDF_x.fit_transform(X_train).toarray()
X_testT = TFIDF_x.transform(X_test).toarray()

print(X_trainT.shape)
print(X_testT.shape)
```

```
#Euclidian
save_distances(X_testT, c, kernel, distance.euclidean, "Euclidian", "T", currentFolder)
```

```
#BrayCurtis
save_distances(X_testT, c, kernel, distance.braycurtis, "BrayCurtis", "T", currentFolder)
```

```
#Canberra
save_distances(X_testT, c, kernel, distance.canberra, "Canberra", "T", currentFolder)
```

```
#Manhattan
save_distances(X_testT, c, kernel, distance.cityblock, "Manhattan", "T", currentFolder)
```

```
X_train = Count_x.fit_transform(X_train).toarray()
X_test = Count_x.transform(X_test).toarray()

print(X_train.shape)
print(X_test.shape)
```

```
#Gaussian
X_trainG = VectGaussian_NB.fit_transform(X_train, y_train)
X_testG = VectGaussian_NB.transform(X_test)

print("(", len(X_trainG), len(X_trainG[0]), ")")
print("(", len(X_testG), len(X_testG[0]), ")")
```

```
#Euclidian
save_distances(X_testG, c, kernel, distance.euclidean, "Euclidian", "G", currentFolder)
```

```
#BrayCurtis
save_distances(X_testG, c, kernel, distance.braycurtis, "BrayCurtis", "G", currentFolder)
```

```
#Manhattan
save_distances(X_testG, c, kernel, distance.cityblock, "Manhattan", "G", currentFolder)
```
