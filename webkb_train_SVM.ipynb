```
def svm_serializer(X, y, c, kernel):
  arr =[]
  for k in kernel:
    temp = []
    for i in c:
      clf_SVC = SVC(C=i, kernel=k)
      clf_SVC.fit(X, y)
      filename = currentFolder + 'models/CRN-' + k + "-" + str(i) + ".joblib"
      dump(clf_SVC, open(filename, 'wb'))
      #print("saved ", filename, end="\t")
      temp.append(clf_SVC)
    arr.append(temp)
  return arr
```
